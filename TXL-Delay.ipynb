```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
import os
import plotly.express as px
from datetime import datetime
```

```python
data = pd.read_csv("FLIGHT.csv")
delay=data.copy()
delay.columns
```

```python
delay.shape
```

```python
col_drop=['DAY_OF_WEEK','FLIGHT_NUMBER','TAIL_NUMBER','DIVERTED','CANCELLED','CANCELLATION_REASON','WHEELS_OFF','WHEELS_ON']
delay.drop(col_drop, axis=1, inplace=True)
delay.head(5)
```

```python
delay.describe()
```

```python
delay.info()
```

```python
delay.dropna(subset=['DEPARTURE_TIME', 'ARRIVAL_TIME', 'ELAPSED_TIME'],inplace = True)
```

```python
cl=['SCHEDULED_DEPARTURE','DEPARTURE_TIME','ARRIVAL_TIME','SCHEDULED_ARRIVAL']
for x in cl:
    delay[x] = delay[x].astype(int)
    delay[x] = (delay[x]//100).apply(lambda a: "{}{}".format(a,':') if ((a-10)>=0) else "{}{}{}".format('0',a,':')) + (delay[x]%100).apply(lambda a: "{}".format(a) if ((a-10)>=0) else "{}{}".format('0',a))
    delay.loc[delay[x]=='24:00',x]='00:00'
    delay[x] = pd.to_datetime(delay[x], format='%H:%M')
    delay[x]=delay[x].apply(lambda c : datetime.strftime(c,"%H:%M"))
```

```python
delay['DATE'] = delay['DAY'].apply(lambda x: "{}{}".format(x,'/')) + delay['MONTH'].astype(str)
delay['DATE'] = delay['DATE'].apply(lambda x: "{}{}".format(x,'/')) + delay['YEAR'].astype(str)
delay['DATE'] = pd.to_datetime(delay.DATE, format='%d/%m/%Y')
delay['DATE'] = delay['DATE'].apply(lambda x : datetime.strftime(x,"%Y-%m-%d"))
delay['DATE']
```
